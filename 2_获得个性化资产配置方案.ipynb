### **Step 2：生成个性化资产配置方案**

现在，环境已经就绪。让我们来调用 Gemini，让它扮演一位专业的理财顾问。

我们会向 Gemini 提供一个用户的基本情况（年龄、月结余、投资目标和风险偏好），然后让它生成一份详细的资产配置方案。

**💡 思考一下**：你可以修改 `user_profile` 中的信息，换成你自己的情况，看看AI会给你什么样的建议。

👇 **运行下方代码，获取你的专属理财规划。**

```python
# 初始化 Gemini Pro 模型
model = genai.GenerativeModel("gemini-2.5-flash-lite")

# 定义用户画像 (Profile)
# 你可以修改这里的内容，定制为你自己的情况
user_profile = """
年龄：32岁
职业：互联网从业者
月结余：5000元
投资目标：5年内积攒一套房子的首付（约50万元）
风险偏好：中等，能接受一定的市场波动以换取更高回报
"""

# 创建一个 Prompt (提示)
prompt = f"""
作为一名专业的理财规划师，请根据以下用户信息，为他/她设计一份详细的月度资产配置方案。
方案需要包含股票、基金、定存的具体投资比例，并用通俗易懂的语言解释为什么这么分配。

---
用户信息：
{user_profile}
---

请以Markdown格式输出，重点内容加粗。
"""

# 调用 Gemini API 生成内容
print("🧠 正在请求 Gemini 生成资产配置方案...")
response = model.generate_content(prompt)

# 使用 Markdown 格式美化输出
print("\n============== AI 理财顾问的建议 ==============")
display(Markdown(response.text))
```

### **Step 3：收益计算与未来财富预测**

有了AI给出的资产配置比例，我们最关心的问题是：**“按照这个计划，5年后我能有多少钱？”**

接下来，我们将用 Python 代码来模拟这个过程。我们会基于AI给出的配置比例，结合一个合理的年化收益率假设，通过**复利计算**来预测未来的资产增长。

**核心概念：复利**
复利就是“利滚利”，每一期的利息都会并入本金，成为下一期计算利息的基础。时间越长，复利效应越明显。

```python

# --- Configuration ---
config = {"股票": 0.40, "基金": 0.40, "定存": 0.20}
rates = {"股票": 0.09, "基金": 0.06, "定存": 0.025}
monthly_investment = 5000
investment_period_years = 5
months = investment_period_years * 12

# --- Simulation ---
principal = 0
records = []
weighted_monthly_rate = sum(config[asset] * (rates[asset] / 12) for asset in config)
for m in range(1, months + 1):
    principal = (principal + monthly_investment) * (1 + weighted_monthly_rate)
    records.append(principal)

# --- Results ---
df = pd.DataFrame(records, columns=["累计资产"], index=range(1, months + 1))
df.index.name = "月份"
final_amount = df['累计资产'].iloc[-1]
print(f"\n📈 按照此计划投资 {investment_period_years} 年 ({months} 个月)，")
print(f"预计最终资产可达到：¥{final_amount:,.2f} 元")
print("\n--- 每年度末资产情况 ---")
df_display = df.iloc[11::12].copy()
df_display.columns = ["累计资产 (元)"]
print(df_display)

# --- Plotting ---
plt.figure(figsize=(12, 7))
plt.plot(df.index, df['累计资产'], marker='o', linestyle='-', markevery=12)
plt.title(f'Wealth Growth Curve over {investment_period_years} Years', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Assets (¥)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### **Step 4：策略动态调整与智能提醒**

市场是不断变化的，一个好的理财计划也需要动态调整。

假设一年后，我们发现整体的年化收益率没有达到预期目标。这时，我们可以再次求助于AI，让它给我们一些调整建议。

👇 **运行下方代码，模拟一次策略调整。**

```python
# 假设我们的目标是实现 7% 的年化收益率
target_yield = 0.07
# 但经过一年，我们计算出实际的年化收益率只有 5%
current_yield = 0.05

# 获取原始配置方案的文本
original_plan_text = response.text

# 创建一个新的调整建议 Prompt
adjustment_prompt = f"""
作为一名专业的理财规划师，我当前的资产配置方案如下：
{original_plan_text}

我的目标年化收益率是 **{target_yield*100:.1f}%**，但目前实际的年化收益率只有 **{current_yield*100:.1f}%**。
请基于这个情况，为我提供一些具体的调整建议，以帮助我更接近目标收益率。
同时，请为我生成一句精炼的提醒语，用于记录在我的投资日志中。
"""

print("🧠 正在请求 Gemini 生成策略调整建议...")
adjustment_response = model.generate_content(adjustment_prompt)

print("\n============== AI 策略调整建议 ==============")
display(Markdown(adjustment_response.text))
```
